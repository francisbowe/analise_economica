```python
#funcao para coverter os dados em valor absoluto
def autopct_format(values):
  def my_format(pct):
    total = sum(values)
    val = int(np.round(pct * total / 100))
    return f'{pct:.1f}%\n({val})'
  return my_format
```

```python
maior_valor = df['logan_passengers'].max()
menor_valor = df['logan_passengers'].min()

print("Maior valor:", maior_valor)
print("Menor valor:", menor_valor)
```

```python
# Gráfico de linha com três indicadores
# Evolução dos Passageiros
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['logan_passengers'], label='Logan Passengers', color='blue')
plt.title('Evolução dos Passageiros em Logan')
plt.xlabel('Ano')
plt.ylabel('Passageiros')
plt.grid(True)
plt.show()

# Evolução da Taxa de Ocupação de Hotéis
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['hotel_occup_rate'], label='Taxa de Ocupação', color='orange')
plt.title('Evolução da Taxa de Ocupação dos Hotéis')
plt.xlabel('Ano')
plt.ylabel('Taxa de Ocupação')
plt.grid(True)
plt.show()

# Evolução do Valor Médio Diário dos Hotéis
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['hotel_avg_daily_rate'], label='Valor Médio Diário', color='green')
plt.title('Evolução do Valor Médio Diário dos Hotéis')
plt.xlabel('Ano')
plt.ylabel('Valor em Dólares')
plt.grid(True)
plt.show()

```

```python
#criando um grafico entre correlacao de dois indicadores
plt.figure(figsize=(14, 6))

# Linha 1: Preço Médio Diário dos Hotéis
plt.plot(df['Year'], df['hotel_avg_daily_rate'], label='Preço Médio Diário dos Hotéis')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.title('Preco Medio Diário dos Hotéis')
plt.legend()
plt.grid(True)
plt.show()

# Linha 2: Taxa de Desemprego
plt.plot(df['Year'], df['unemp_rate'], label='Taxa de Desemprego')

plt.xlabel('Ano')
plt.ylabel('Valor')
plt.title('Taxa de desemprego Desemprego')
plt.legend()
plt.grid(True)
plt.show()
```

Análise de Correlação (Mapa de Calor)

```python

# Primeiro, cria uma matriz de correlação
corr = df.corr()

# Agora, cria o mapa de calor
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Correlação dos Indicadores Econômicos')
plt.show()
```
